```python
## First lets us make date,month,year seperate colums so that helps in ml model
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
## Drop Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
## still date month and years are in Object type So convert them to numerical for model


df['Date'] = df.Date.astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
## Arrival time 

df["Arrival_hour"] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[0]


# drop arrival time
df.drop('Arrival_Time',inplace=True,axis=1)
```

```python
df['Arrival_minute']
```

```python
df.head(2)
```

```python
# Departure time
df['Dep_hour'] = df['Dep_Time'].str.split(":").str[0] 
df['Dep_minute'] = df['Dep_Time'].str.split(":").str[1] 


#drop Dep_time

df.drop('Dep_Time',axis=1,inplace=True)
```
