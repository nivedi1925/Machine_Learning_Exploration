```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
## Value count return how many unique vales repeate in the column
print(df.quality.value_counts())


##Visualize
df.quality.value_counts().plot(kind='pie')
plt.show()
```

```python
### Here conclution is that it is an imbalanced dataset
```

### Note:
If there is an imbalance dataset especially in classification problmes use <br>
1. Upsampleing -- but this simply makes more point in the already existing point<br>
2. SMOTE -- this creates line between nearer points and creates data points along the line

```python
## distribution of each column in the given column
plt.figure(figsize=(20,5))
for column in df.columns:
    sns.histplot(df[column],kde=True)
```

```python
## distribution of each column in the given column

for column in df.columns:
    plt.figure(figsize=(20,5))
    sns.histplot(df[column],kde=True,color='yellowgreen')
    plt.show()
```

```python
## iniveriate,bivariate,multivariate analysis
sns.pairplot(df)
plt.show()
```

```python
### categorical plot
sns.catplot(x='quality',y='alcohol',data=df,kind="box")
```
