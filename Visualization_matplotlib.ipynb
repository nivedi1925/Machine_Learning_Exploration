## Data Visualization using *matplotlib* 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [5,3,8,9,4,3,1,3,6]
df = pd.DataFrame({"x":x,"y":y})
```

```python
df.head(2)
```

```python
## Simple plot
plt.plot(df["x"],df.y,color="red",marker="X",markersize='5',linewidth=2,linestyle='--')
plt.title("Sample data set with plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
### simpler version --> directly takes df and draws line graph for all columns
df.plot(kind='hist')
```

```python
plt.hist(df.y,bins=3,color=["red"])
plt.show()
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind="area")
plt.title("Sample")
plt.show()
```

```python
## Back to using plt 
#Scatter plot
```

```python
plt.scatter(df.x,df.y,color=['green'],marker='*',edgecolors='red',linewidths=1,s=5)
plt.show()
```
